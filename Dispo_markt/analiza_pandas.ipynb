<h1> Analiza transportnih podatkov </h1>

<h2> Pregled in urejanje podatkov</h2>

```python
# naložimo Pandas 
import pandas as pd
import numpy as np
# import os.path

# %matplotlib notebook
# Se ne bom poglabljal v grafične predstavitve podatkov.

#Uvoz podatkov-----------
pathData = "data_set - transported.csv"
data = pd.read_csv(pathData)
#------------------------

# Izpisovanje -----------
pd.options.display.max_rows = 30
pd.options.display.max_columns = len(data.columns)
#------------------------


# Številske stolpce Str -> Int
numberColumns = ['Deliveries', 'Attachments', 'Unique destinations']

for i in numberColumns:
    data[i] = data[i].astype(int)

# Weight stolpec Str -> float

data['Weight'] = data['Weight'].apply(lambda x: float(x.replace(',','')))

#------------------------

# data.columns = data.columns.str.replace(' ', '')
```

```python
data
```

<h3> Različne vrednosti </h3>

<p>Pogledamo koliko je različnih vrednosti za posamezni stolpec. </p>

```python
#Koliko je različnih vrednosti za posamezen stolpec?

uniqueDataDic = {}
uniqueDataDicCount = {}

for type in data.columns[1:]:
    x = data[type].unique().tolist()
    uniqueDataDic[type] = x
    uniqueDataDicCount[type] = [len(x)]

uniqueDataCount = pd.DataFrame.from_dict(uniqueDataDicCount)

uniqueDataCount
```

```python
#Poglejmo neočitne različne vnose po stolpcih

notObviousData = ['ETA status','ETA type','Difference with ETA','Deliveries','Attachments','City - Origin','Country - Origin','Country - Destination','Unique destinations','Vehicle (Requirement)']

#City - Destination -> preveč različnih

for type in notObviousData:
    print(f"{type} : {uniqueDataDic[type]}")
```

<p> Že tukaj lahko vidimo neke osnovne podatke o delovanju podjetja.</p>


<h3>Datum</h3>

<p>Uredimo datum. Dani zapis %Y-%m-%d ločimo, dodamo stolpce za mesec in dan.</p>

```python
# Iz datuma dodamo določimo mesec, dan


#Loading date -> v datetime zapis
data['Loading date'] = pd.to_datetime(data['Loading date'],format='%Y-%m-%d')
data['Unloading date'] = pd.to_datetime(data['Unloading date'],format='%Y-%m-%d')

#Dodajanje stolpcev day, month
data['Loading month'] = data['Loading date'].dt.month
data['Loading day'] = data['Loading date'].dt.day

data['Unloading month'] = data['Unloading date'].dt.month
data['Unloading day'] = data['Unloading date'].dt.day


#Časovno trajanje poti

data['Duration (days)'] = (data['Unloading date'] - data['Loading date']).dt.days

data
```

```python
#Za katere mesece so podatki

type = 'Loading month'
print( f"Loading month : {data[type].unique().tolist()}")
```

<h3> Odstranjevanje nepotrebnih podatkov</h3>
 
<p>Poglejmo le uporabne podatke. Odstranimo notUseful stolpce. Glej kodo.</p>

<p>Spodaj so podani urejeni podatki.</p>

```python
notUseful = ['Loading date', 'Unloading date', 'Loading period', 'Unloading period', 'ZIP - Origin', 'Region - Origin', 'ZIP - Destination', 'Region - Destination', 'Quote request type', 'Quote request', 'Quote request currency', 'Quote request comment']

useful = [i for i in data.columns.to_list() if i not in notUseful]

sortedData = data.loc[:, useful]

sortedData
```

<h2> Mesečna analiza</h2>

```python
#Mesečna analiza

months = sortedData["Loading month"].unique().tolist()

dataMonthsDic = {'Months' : months}

sumWeight = []
sumDeliveries = []
countInput = []

for month in months:
    countInput.append(len(sortedData[data['Loading month'] == month]))
    sumWeight.append(sortedData.loc[data['Loading month'] == month, 'Weight'].sum())
    sumDeliveries.append(sortedData.loc[data['Loading month'] == month, 'Deliveries'].sum())

dataMonthsDic['All trips'] = countInput
dataMonthsDic['Sum Weight'] = sumWeight
dataMonthsDic['Sum Deliveries'] = sumDeliveries


dataMonths = pd.DataFrame.from_dict(dataMonthsDic)

dataMonths
```

<p> Pregled začetnih lokacij po posameznih mesecih.</p>

```python
sortedData.groupby(['Loading month', 'City - Origin'])['ID'].count()
```

<h2>Lokacijska analiza</h2>


<h3>Začetna lokacija</h3>

<p>Spodaj je tabela vsote teže in števila izvozov iz posameznega mesta.</p>

```python
# Izvoz 

sortedData.groupby('City - Origin')['Weight'].agg(['sum','count'])
```

<h3>Končna lokacija</h3>

<p>Tabela podatkov grupiranih po končni destinaciji s številom prevozov in vsoto teže.</p>

```python
#uvoz

sortedData.groupby('City - Destination')['Weight'].agg(['sum','count']).sort_values(by=['count', 'sum'])
```

<h3>Mednarodni prevozi</h3>



```python
# International trips

sortedData['International'] = np.where(sortedData['Country - Origin'] == sortedData['Country - Destination'], True, False)

sortedData.groupby('International')['Weight'].agg(['sum','count'])
```

<h2>Zamude</h2>

<p>Poglejmo prevoze gleda na ETA status.</p>
<p>Spodaj so izpisani le primeri ko je ETA status različen od None. Takih primerov je le malo, ne moremo kaj več sklepati, podatki so nepopolni.</p>

```python
#Zamude 
#ETA status : ['None', 'On time', 'Delayed']
statusData = sortedData.loc[sortedData['ETA status'] != 'None']
statusData
```

<h2>Tipi Tovornjakov</h2>

<p> Vsota prevožene teže, števila prevozov in povprečne teže na vrsto tovornjaka.</p>

```python
#Tipi tovornjakov

truckData = sortedData.groupby(['Vehicle (Requirement)'])['Weight'].agg(['sum','count']).sort_values(by=['count'])

truckData['Average weight'] = truckData['sum']/truckData['count']

truckData
```
